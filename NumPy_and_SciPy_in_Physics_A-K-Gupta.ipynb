## First and second derivative using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the points in x array i.e., this is the x-data.
x = np.linspace(-5, 5, 100)
# print(x)

# The y-data is generated using the following function.
y = 1 / (1 + np.exp(-x))
# print(y)

# Calculate the first derivative: z = dy/dx. The np.diff() function computes the difference between successive elements in a numpy array. np.diff() generates an array whose length is one less than the original array.
z = np.diff(y) / np.diff(x)
# print(z)

# Plot of y vs x
plt.plot(x, y)
plt.show()

# plot of z or dy/dx vs x
plt.plot(x[:-1], z)
plt.show()

# See that the maximum in the 2nd plot is at the inflexion point of the 1st plot.
```

```python
t = np.linspace(0, 10, 100)  # t array has 100 elements.
# print(t)

x = 1 / (1 + np.exp(-2*t + 8))  # x array has 100 elements.
# print(x)

# Velocity, v = dx/dt.
v = np.diff(x) / np.diff(t)  # v array has one less element than the original array. So, it contains 99 elements.

# Acceleration, a = dv/dt = d2x/dt2.
a = np.diff(v) / np.diff(t[:-1])  # Here v array has 99 elements and t array has 100 elements. So, to avoid the length problem, we remove the last element of t array using indexing t[:-1], which takes all the elements except the last element. So, now v and t[:-1] both arrays have 99 elements. And we take difference on these. After taking the np.diff(), of course 'a' array has 98 elements.

# Plot x vs t.
plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x')
plt.show()

# Plot of v vs t.
plt.plot(t[:-1], v)  # Here v and t[:-1] both have 99 elements.
plt.xlabel('t')
plt.ylabel('v')
plt.show()

# Plot of a vs t.
plt.plot(t[1:-1], a)  # Here both a and t[1:-1] both have 98 elements. We have discarded the first and last elements of t array to avoid the length problem in plotting. We could also use t[:-2] (in this case, the last two elements are removed from the plot) instead of t[1:-1].
plt.xlabel('t')
plt.ylabel('a')
plt.show()
```

**Tensor product**: It is different from the Hadamar product. If two tensors A and B have dimension p and q, respectively, then the resulting tensor product, C = A $\otimes$ B, will have the dimesion (p + q). The tensor product C will contain the elements where each element of A will multiply the tensor B.

In Numpy, tensor product can be done using **np.tensordot()** function. For usual tensor product, use **axes = 0**.



```python
A = np.array([1, 2, 3])  # 1D numpy array.
B = np.array([4, 5])  # 1D numpy array.

C = np.tensordot(A, B, axes = 0)
print('The tensor product of A and B is = ', C)

# Notice that C is a 2D numpy array.

# The tensor product is used in neural network.
```

```python
# Solve linear equations using Numpy.
# x1 + 2x2 - x3 = 1
# 2x1 + x2 + 4x3 = 2
# 3x1 + 3x2 + 4x3 = 1

# The above set of equations can be written in the matrix form: AX = B.
# where, A = [[1, 2, -1], [2, 1, 4], [3, 3, 4]] is the 2D array or matrix of the coefficients.
# And X is the 1D array or vector [x1, x2, x3]. B = [1, 2, 1] is the vector of the RHS numbers.

# Then, X = A^-1 B.

A = np.array([[1, 2, -1],
              [2, 1, 4],
              [3, 3, 4]])

B = np.array([1, 2, 1])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)
print(X)

# So, value of x1 = 7, x2 = -4, x3 = -2.

# Direct solve.

np.linalg.solve(A, B)
```

```python
# Eigenvalue problem. Eigenvalues and eigenvectors.

# We use np.linalg.eig() function to calculate eigenvalues and eigenvectors of a matrix or higher dimensional numpy arrays.

A = np.array([[1, 2],
              [3, 4]])

print(np.linalg.eig(A))

# The eig() function returns a tuple of arrays. The 1st array is the collection of the eigenvalues. And the 2nd array is the
# collection of the eigenvectors. The two columns of the 2nd array are the two eigenvectors.
```

```python
# Eigenvalues and eigenvectors unpacked.

# Unpacking the arrays of eigenvalues and eigenvectors.
u, v = np.linalg.eig(A)

print(u)
print(v)

# Unpacking the eigenvalues.
u1, u2 = u
print(u1, u2)

# Unpacking the eigenvectors. Here, 1st column is the 1st eigenvector and 2nd column is the 2nd eigenvecor.
v1 = v[:, 0]
v2 = v[:, 1]
print(v1)
print(v2)
```

```python
# Monte Carlo estimate of pi.

# Here we generate random numbers in a square region of length -1 to 1 for x and y coordinates. So the side length of
# this square is = 2. Area of this square = 2**2 = 4. Now we draw a circle of radius 1 within this sqaure such that the
# circle touches the sides of the square. So, area of this circle is = pi * 1**2 = pi. So, the ratio of the area of circle
# and area of square = pi / 4.
# Now we generate large number of random numbers uniformly within the square region. So, a lot of random numbers will fall
# insdie the circle region. When the random numbers are large, the ratio of the random numbers within the circle and
# total random numbers will be equal to the ratio of their areas.
# In this way, we count the number of random numbers that fall inside the circle, and take the ratio of it with the total
# random numbers. This is = pi / 4. So, pi = 4 * the above ratio. This is how we can compute the value of pi. This is the
# way how Monte Carlo simulations works.

x = np.random.uniform(-1, 1, 100000)
y = np.random.uniform(-1, 1, 100000)

z = x**2 + y**2

u = z[z<1]

print(len(z), len(u))

print(4 * (len(u) / len(z)))
```

Consider the Gaussian integral:

$\frac{1}{\sigma \sqrt{2 \pi}} \int_{-}^{} e^{-(x - \mu)^2 / 2\sigma^2} dx = 1$

```python
# About infinity in numpy.

import numpy as np

print(np.inf)
print(type(np.inf))

print(9999999999999999999999999999999999 > np.inf)
print(-99999999999999999999999999999999 < -np.inf)

print(np.inf + np.inf)
print(np.inf * 10)
print(np.inf * np.inf)
print(np.inf - np.inf)
print(np.inf / np.inf)
```

```python
import numpy as np

from scipy.integrate import quad, simps

# Define the mu and sigma of the Gaussian function i.e., mean or peak of the Gaussian and standard deviation or spread
# of the Gaussian function.
mu, sigma = 0.5, 1.0

# Define the Gaussian function.
f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp((-(x - mu)**2) / (2 * sigma**2))

# Integration of the Gaussian fucntion using scipy quad() function, in the range -infinity to +infinity.
print(quad(f, -np.inf, np.inf))

# It returns a tuple (result, error). So, the result is 1 which matches with the exact analytical result.
```

```python
# Now take smaller values of sigma, and check the result of the integration.

mu, sigma = 0.5, 0.1
print(quad(f, -np.inf, np.inf))

mu, sigma = 0.5, 0.01
print(quad(f, -np.inf, np.inf))

mu, sigma = 0.5, 0.005
print(quad(f, -np.inf, np.inf))

mu, sigma = 0.5, 0.001
print(quad(f, -np.inf, np.inf))

mu, sigma = 0.5, 0.0001
print(quad(f, -np.inf, np.inf))
```

**When Gaussian tends to the delta function**

**Note:** A Gaussian function narrows down systemetically, with smaller and smaller $\sigma$ values. In the limit $\sigma \rightarrow 0$, this becomes a delta function $\delta(x-\mu)$, where $\mu$ is the mean of the Gaussian distribution.

Notice the sequence of integration results above. If the standard deviation $\sigma$ is very small (i.e., width of the curve is very small) and the domain of integration is very large, the integration misses the dominant contributions (area under the narrow curve in the central region). So, instead of searching for the infinite extent on the $x$-axis, we must concentrate on the central part from which the dominant contribution comes. Now, the question is what will be range of integration?

We may integrate over the region which extends upto a few $\sigma$ distances on either of the peak $x = \mu$. e.g., we can integrate from $\mu - 5\sigma$ to $\mu + 5\sigma$. Check the integration result and adjust this range.

```python
import numpy as np

from scipy.integrate import quad, simps

# Define the mu and sigma of the Gaussian function i.e., mean or peak of the Gaussian and standard deviation or spread
# of the Gaussian function.
mu, sigma = 0.5, 0.001

# Define the Gaussian function.
f = lambda x: (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp((-(x - mu)**2) / (2 * sigma**2))

# Integration of the Gaussian fucntion using scipy quad() function, in the range -infinity to +infinity.
print(quad(f, mu - 6*sigma, mu + 6*sigma))

# It returns a tuple (result, error). So, the result is 1 which matches with the exact analytical result.

# This integration can also be done using 'simps' method.
```

### Fresnel integral

The Fresnel integral:

\begin{equation}
    \int_{0}^{} sin(x^2)~dx =  \sqrt{\frac{\pi}{8}}
\end{equation}

```python
# Plot of the integrand.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 1000)
f = lambda x: np.sin(x**2)

plt.plot(x, f(x))
plt.show()
```

So, this function has ever decreasin period. As for the narrow Gaussian we took care of the integration range, here, for upper limit we have to take care, since as $x$ increases, the function becomes narrower.

Here, we progressively increase the upper limit of integration inside a loop.

```python
import numpy as np
from scipy.integrate import simps

f = lambda x: np.sin(x**2)

# Define the array of upper limits.
UL = np.arange(1, 100, 0.1)

# Define the empty list of integration result.
I = []

# Run the loop of the upper limit and calculate the integration.
for n in UL:
    x = np.linspace(0, n, 10000)
    I.append(simps(f(x), x))


# For plot of Integration result vs the upper limit.

import matplotlib.pyplot as plt

plt.xlabel('Upper limit')
plt.ylabel('Integration result')
plt.axhline(0.6267, color = 'black')  # Draw a horizontal line at the actual value of the integration.
plt.plot(UL, I)
plt.show()
```

The value of the integration result varies around 0.6267 and converges toward this value as we increase the upper limit.

```python
# Let's calculate the mean of the integration results by discarding the first 500 values as they oscillate more.

print(np.mean(I[500:]))
```

**Cornu's spiral**

\begin{equation}
    C^2 + S^2 = \left( \int_{0}^{t} cos(\frac{1}{2} \pi x^2) \right)^2 + \left( \int_{0}^{t} sin(\frac{1}{2} \pi x^2) \right)^2
\end{equation}

This is a graph of $S$ vs $C$. This estimate the intensity in terms of the arc length $x$.

```python
# Cornu's spiral: It is a graphical presentation to compute and predict the Fresnel diffraction pattern from
# various simple obstacles.

import numpy as np
from scipy.integrate import simps

UL = np.arange(-50, 50, 0.01)

f1 = lambda x: np.cos(np.pi * x**2 * 0.5)

f2 = lambda x: np.sin(np.pi * x**2 * 0.5)

C = []
S = []
for i in UL:
    x = np.linspace(0, i, 10000)
    C.append(simps(f1(x), x))
    S.append(simps(f2(x), x))

C = np.array(C)
S = np.array(S)

# For plot.

plt.xlabel('C')
plt.ylabel('S')
plt.plot(C, S)
plt.show()
```

## Double integral using dblquad() function

In the function $dblquad(f(x, y), a, b, gfun(x), hfun(x))$, $f(x, y)$ is the function, $a$, $b$ are the $x$ limits, and $gfun(x)$, $hfun(x)$ are the $y$ limits. Note that the $y$ limits are functions here. So we give them directly in the argument of $dblquad()$ as $lambda$ function.

1. Area of circle: $x^2 + y^2 = 36$.

**Ans:** We have to calculate $\int \int f(x)~dxdy$.

If $x^2 + y^2 = a^2$, then $y = \pm \sqrt{a^2 - x^2}$.

So, the limit of integration will be $x = [-a, a]$ and $y = [-\sqrt{a^2 - x^2}, +\sqrt{a^2 - x^2}]$.

Now, in the problem-1, our $f(x) = 1$. And in the problem-2, our $f(x) = \sqrt{x^2 + y^2}$.

```python
import numpy as np
from scipy.integrate import dblquad

f = lambda x, y: 1

print(dblquad(f, -6, 6, lambda x: -np.sqrt(36 - x**2), lambda x: np.sqrt(36 - x**2)))

# Here look at how the limit of x and y are given. x ranges from [-6, 6], and y ranges are defined in terms of the lambda
# function with limits -sqrt{a^2 - x^2} to +sqrt{a^2 - x^2}. Note that the integration limits of y are functions itself.
```

## Ordinary differential equations (ODE)

For solving ODE, we use **odeint()** function of the **integrate** module of **Scipy** library.

```python
import numpy as np
from scipy.integrate import odeint

help(odeint)
```

Of the first 3 arguments, **func** is the name of the function to be integrated. $y0$ is the initial value of the dependent variable (let's say $y = f(t)$, where $y$ or $f(t)$ is the dependent variable and $t$ is the independent variable). And $t$ is the array of the independent vatiable over the range in which the solution has to be found. There are other keyword arguments which can be used to get the solutiion in desired ways.

For higher order ODEs, where we have more than one function to be integrated, we need to pack the functions under one name in order to put as the first argument **func**. Similarly more than one dependent variables are to be packed up under one name to put as the second argument $y0$.

### First order ODE

The general form of first order ODE:

\begin{equation}
\frac{dx}{dt} = f(x, t)
\end{equation}

To solve this for $x(t)$, we need to have one initial condition. For example, $x = x_0$, at $t = t_0$. The **odeint()** function takes the function name ($f$) as argument along with the initial value (here the 2nd argument $y0 = x_0$) for the dependent variable, and the array of the independent variable ($t$) over which the the solution is to be found.

#### Example: Radioactive decay

Consider the first order ODE of radioactive decay of nuclear mass:

\begin{equation}
\frac{dx}{dt} = -kx
\end{equation}

Here, $f(x, t) = -kx$. Here $x(t)$ is the dependent variable and $t$ is the independent variable.

```python
import numpy as np
from scipy.integrate import odeint

# Define the function.

def f(x, t):
    return -k*x

k = 1.0  # parameter

# Initial value of dependent variable.
x0 = 100

# Array of independent variable.

t = np.linspace(0, 10, 100)

# Apply odeint() function to get the solution.

sol = odeint(f, x0, t)
print(len(sol))
print(sol)

# See that the first value in the solution array is 100 i.e., the initial value.
```

It reurns an array i.e., $x(t)$ array. Basically it calculates $x$ at each value of $t$ of the $t$ array. This array contains column of values of $x$ at all points in the given array $t$. The function **odeint(f, x0, t)** takes 3 default arguments, where $f$ is the user defined function name that contains the derivative $\frac{dx}{dt}$. Integration is done for $x$ at all points of the $t$ array, where $x0$ is the initial value. $x0 = 100$ means, $x = 100$ at the initial value of the $t$ array, i.e., at $t = 0$.

Note the order of the arguments: **func(dependent variable, independent variable)** (i.e., $f(x, t)$, and not $f(t, x)$).

```python
# What is about the solution array?

print(type(sol))  # Solution is a numpy array.

print(sol.ndim)  # 2D numpy array.
print(sol.shape)  # Solution is a 2D array of size (100x1) i.e., it has 100 rows and 1 column.
```

### Second order ODE

We can split a second order ODE into two coupled first order ODEs. For each of the first order ODE, we can use **odeint()** funcion.

#### Example: Damped harmonic oscillation

Consider damped harmonic oscillation:

\begin{equation}
\frac{d^2x}{dt^2} + \lambda \frac{dx}{dt} + kx = 0
\end{equation}

We recast the above second order ODE into following two first order ODEs:

\begin{equation}
\frac{dx}{dt} = y = f(x, y, t) \\
\frac{dy}{dt} = -\lambda y - kx = g(x, y, t)
\end{equation}

Note the order of dependent variables in $f(x, y, t)$ and not $f(y, x, t)$, and same for $g(x, y, t)$. i.e., **func(dependent variable which was initially in the eqaution, dependent variable which you introduce, independent variable)**.

To solve we need to initial values, one is for $x$ and one is for $y$, and the array of independent variable $t$.

As the function **odeint(func, y0, t)** can take only 3 default arguments, we pack two functions $f$ and $g$ into one name to put in place of $func$. Similarly we pack two initial values $(x0, y0)$ into one name to put in place of $y0$.

```python
import numpy as np
from scipy.integrate import odeint

def damp(u, t):
    x, y = u   # unpacking u.
    f = y
    g = -k*x - lam * y
    return [f, g]    # return two functions in pack.

# Parameters.
k = 1
lam = 0.4

# Initial values.
x0 = 0
y0 = 1
u0 = [x0, y0]  # packing.

# Array of the independent variable.
t = np.linspace(0, 20, 100)

# Calculate the solution using odeint().

sol = odeint(damp, u0, t)
print(sol)

# It returns a 2D numpy array. Number of rows = length of t (independent variable) array. Number of columns = order of the ODE.
# The first column is the solution for x and the second column is the solution for y = dx/dt.

# We define the function 'damp' which returns 2 functions f and g, in pack. The two dependent variables x and y are
# packed as u = [x, y], and they are again unpacked inside the 'damp' function. Initial values, u0 = [x0, y0] = [0, 1].

# See here, for a 2nd order ODE, we have 2 dependent variables x and y, and 2 functions f and g. And we will have
# 2 first order ODEs for each of the dependent variables.
```

```python
print(sol.shape)

# The solution has 100 rows (associated with the length of the independent variable array) and 2 columns,
# one is for x and one is for y.
```

```python
# Print x and y solutions separately i.e., x is the 0th column of the sol array, and y is the 1st column of the sol array.

x = sol[:, 0]
y = sol[:, 1]

print('The x array: ', x)
print('The y array: ', y)
```

#### Example: Forced vibration

Consider damped harmonic oscillation:

\begin{equation}
\frac{d^2x}{dt^2} + \lambda \frac{dx}{dt} + kx = F\cos(\omega t)
\end{equation}

We recast the above second order ODE into following two first order ODEs:

\begin{equation}
\frac{dx}{dt} = y = f(x, y, t) \\
\frac{dy}{dt} = F\cos(\omega t) -\lambda y - kx = g(x, y, t)
\end{equation}

From the previous problem, we have only $F\cos(\omega t)$ term extra here.

Amplitude of the steady state, $A(\omega) = \frac{F}{\sqrt{(k - \omega^2)^2 + \lambda^2 \omega^2}}$ (exact solution).

```python
import numpy as np
from scipy.integrate import odeint

def forced(u, t):
    x, y = u   # unpacking u.
    f = y
    g = F * np.cos(w * t) -k*x - lam * y
    return [f, g]    # return two functions in pack.

# Parameters.
k = 5
lam = 0.5
F = 1.0

# Initial values.
x0 = 0
y0 = 1
u0 = [x0, y0]  # packing.

# Array of the independent variable.
t = np.arange(0, 100, 0.1)

# Frequency range.
freq = np.arange(0.5, 5, 0.02)

A = []
V = []

# Calculate the solution using odeint().

for w in freq:
    sol = odeint(forced, u0, t)
    x = sol[:, 0]
    y = sol[:, 1]
    amp = np.max(x[-400:])  # Here initial points are discarded as the system takes some time to settle to equilibrium against the external force. This is very important. Otherwise plot will not match with the exact formula.
    amp1 = np.max(y[-400:])
    A.append(amp)
    V.append(amp1)

# Exact formula.
exact = F / np.sqrt((k - freq**2)**2 + lam**2 * freq**2)

# For plotting.

import matplotlib.pyplot as plt

plt.plot(freq, A, 'o')  # Amplitude calcuated numerically vs freq.
plt.plot(freq, exact, '-')   # exact amplitude vs freq.
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Next we plot position vs time to observe the time evolution and to make an idea about equilibrium to occur.

plt.plot(t, x)
plt.xlabel('Time, t')
plt.ylabel('Position, x')
plt.show()

# And understand why we discarded initial points of x to get the amplitude.
```

#### Exercise-3:

Solve the following set of three coupled first order ODEs:

\begin{equation}
\frac{dx}{dt} = \sigma (y - x), \frac{dy}{dt} = x(\rho - z), \frac{dz}{dt} = xy - \beta z .
\end{equation}

Solving these equations we get the famous Lorentz curves (in the area of chaos). The values of the parameters that Lorentz used; $\sigma = 10, \rho = 28, \beta = 8/3$.

The phase plots (between $x$ and $y$ or $y$ and $z$ etc.) are called Lorentz attractors.

```python
# Define the function.

def lorentz(u, t):
    x, y, z = u
    f1 = sig * (y - x)
    f2 = x * (rho - z)
    f3 = x * y - beta * z
    return [f1, f2, f3]

# Parameters.
sig, rho, beta = 10, 28, 8/3

# Initial values.
x0, y0, z0 = 1, 1, 1
u0 = [x0, y0, z0]

# Array of independent variable.
t = np.arange(0, 100, 0.1)

# Find the solution.
sol = odeint(lorentz, u0, t)
print(sol)

x = sol[:, 0]  # The first column of sol array is x.
y = sol[:, 1]  # The 2nd column of sol array is y.
z = sol[:, 2]  # The 3rd column of sol array is z.

# For plotting.

plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('z')
plt.show()
```

**Application:** 1D linear harmonic oscillator.

The wave function of a 1D LHO:

$ \psi_n(x) = \frac{1}{\sqrt{2^n n!}} (\frac{m \omega}{\pi \hbar})^{1/4} ~ e^\frac{-m \omega x^2}{2 \hbar} H_n \left(\sqrt{\frac{m \omega}{\hbar}} x \right) $ .

```python
# Python script to plot the LHO wave functions.

# We use (m*omega)/hbar = 1.

import numpy as np
from scipy.special import hermite
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 1000)

def psi(n, x):
    Hn = hermite(n)
    prefactor = (1 / np.sqrt(2**n * np.math.factorial(n))) * (1 / (np.pi))**(1/4)
    wave_function = prefactor * np.exp(-x**2 / 2) * Hn(x)
    return wave_function

# Plotting.

for n in range(0, 3):
    plt.subplot(1, 3, n+1)
    plt.plot(x, psi(n, x))
    plt.title('n = %d' %n)
    plt.tight_layout(pad = 0.75)
plt.show()

# Here we plot ground state and two excited states of 1D LHO.
```

## Fourier series

The Fourier series is given by,

$ f(x) = \frac{a_0}{2} + \sum_{n = 1}^{\infty} a_n \cos(nx) + \sum_{n = 1}^{\infty} b_n \sin(nx)$ ,

where,

$a_0 = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x)~dx$, \\

$a_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \cos(nx)~dx$, \\

$b_n = \frac{1}{\pi} \int_{-\pi}^{\pi} f(x) \sin(nx)~dx$ .

Here, we have a Fourier expansion of a periodic function $f(x)$, and the Fourier coefficents are given by the integrals as written.

### Fourier series of sqaure wave

```python
import numpy as np
from scipy.integrate import quad
from scipy.signal import square
import matplotlib.pyplot as plt

# Define the function.

x = np.linspace(-4, 4, 1000)
f = lambda x: square(x)

# quad() function do the integration and returns a tuple, in which the first element is the value of the integral
# and the 2nd element is the error value. We extract here the 1st element.

# Calculate the Fourier coefficients for each n and calculate also Fourier series or the sum of the Fourier terms.

a0 = (1 / np.pi) * quad(f, -np.pi, np.pi)[0]
s = a0 / 2
for n in range(1, 20):
    an = (1 / np.pi) * quad(lambda x: f(x) * np.cos(n*x), -np.pi, np.pi)[0]
    bn = (1 / np.pi) * quad(lambda x: f(x) * np.sin(n*x), -np.pi, np.pi)[0]
    s = s + an * np.cos(n*x) + bn * np.sin(n*x)

# For plotting.

plt.plot(x, s)
plt.plot(x, square(x))
plt.show()
```

### Fourier series of a full-wave rectified signal

```python
# Full wave rectified signal is basically np.abs(sin(x)) curve.

import numpy as np
from scipy.integrate import quad

# Define the function.

x = np.linspace(-6, 6, 1000)
f = lambda x: abs(np.sin(x))

# quad() function do the integration and returns a tuple, in which the first element is the value of the integral
# and the 2nd element is the error value. We extract here the 1st element.

# Calculate the Fourier coefficients for each n and calculate also Fourier series or the sum of the Fourier terms.

a0 = (1 / np.pi) * quad(f, -np.pi, np.pi)[0]
s = a0 / 2
for n in range(1, 20):
    an = (1 / np.pi) * quad(lambda x: f(x) * np.cos(n*x), -np.pi, np.pi)[0]
    bn = (1 / np.pi) * quad(lambda x: f(x) * np.sin(n*x), -np.pi, np.pi)[0]
    s = s + an * np.cos(n*x) + bn * np.sin(n*x)

# For plotting.

plt.plot(x, s)
plt.plot(x, abs(np.sin(x)))
plt.legend(['Fourier sum', 'full-wave rectified signal'])
plt.show()
```

### General period (other than $2\pi$)

So far, we havr the peridicity of $2\pi$ i.e., the periodic function had one complete cycle in the range $[-\pi, \pi]$. \\
Now, for a periodic function having one compelete period or cycle in the range $[-L, L]$, the formula of Fourier seris and the Fourier coefficients changes:

$ f(x) = \frac{a_0}{2} + \sum_{n = 1}^{\infty} a_n \cos(\frac{n \pi x}{L}) + \sum_{n = 1}^{\infty} b_n \sin(\frac{n \pi x}{L})$ ,

where,

$a_0 = \frac{1}{L} \int_{-L}^{L} f(x)~dx$, \\

$a_n = \frac{1}{L} \int_{-L}^{L} f(x) \cos(\frac{n \pi x}{L})~dx$, \\

 $b_n = \frac{1}{L} \int_{-L}^{L} f(x) \sin(\frac{n \pi x}{L})~dx$  .

```python
import numpy as np
from scipy.integrate import quad
from scipy.signal import sawtooth
import matplotlib.pyplot as plt

# Define the function.

L = 8
x = np.linspace(-L, L, 1000)
f = lambda x: sawtooth(x)

# quad() function do the integration and returns a tuple, in which the first element is the value of the integral
# and the 2nd element is the error value. We extract here the 1st element.

# Calculate the Fourier coefficients for each n and calculate also Fourier series or the sum of the Fourier terms.

a0 = (1 / L) * quad(f, -L, L)[0]
s = a0 / 2
for n in range(1, 50):
    an = (1 / L) * quad(lambda x: f(x) * np.cos((n*np.pi*x) / L), -L, L, limit = 100)[0]  # We have increased the value of the limit argument to 100, as we were getting warning with the defalult limit = 50.
    bn = (1 / L) * quad(lambda x: f(x) * np.sin((n*np.pi*x) / L), -L, L, limit = 100)[0]
    s = s + an * np.cos((n*np.pi*x) / L) + bn * np.sin((n*np.pi*x) / L)

# For plotting.

plt.plot(x, s)
plt.plot(x, sawtooth(x))
plt.show()
```

## Discrete Fourier Transform (DFT)

For details derivation and concepts of DFT see A. K. Gupta's book. The Discerete Fourier Transform (DFT) is given by,

**Fourier transform:** $F(k) = \sum_\limits{n=0}^{N-1} f(n) e^{-i \frac{2\pi n k}{N}}$  .

**Inverse Fourier transform:** $f(n) = \frac{1}{N} \sum_\limits{k=0}^{N-1} F(k) e^{i \frac{2\pi n k}{N}}$  .

Here, $n$ and $k$ are the indices of the time sample and frequency sample, respectively (Actually, $n$ is the short form of $t_n = n \Delta t$, and $k$ is the short form of $\omega_k = k \Delta \omega$; where, $\Delta t$ and $\Delta \omega$ are the time interval and frequency interval, respectively.). $N$ is the total number of samples. $f(n)$ is the value of the signal at $t_n$ or $n^{th}$ sample, and $F(k)$ is the fourier transform of the signal. Here we are taking time varying signal $f(n)$ or $f(t_n)$ and performing the Fourier transform of it to get Frequency spectrum $F(k)$ or $F(\omega_k)$. We will have such $N$ frequencies in total as $k$ also varies from $0$ to $N-1$.

In practice, we use further shorter notations: $x_n = f(n)$, and $X_k = F(k)$. So, the abover formula becomes:

**Fourier transform:** $X_k = \sum_\limits{n=0}^{N-1} x_n e^{-i \frac{2\pi n k}{N}}$  ,

where, $N = $ total number of samples, \\
$n = $ current sample, \\
$x_n = $ value of the signal at time $n$ or $t_n$, \\
$k = $ current frequency (it varies from 0 Hz to $N-1$ Hz), \\
$X_k = $ Fourier transformed value which gives the amplitude and phase of the signal.

The periodicity of the Fourier transformed signal can also be shown easily:

$X_{k+N} = \sum_\limits{n=0}^{N-1} x_n e^{-i \frac{2\pi n (k+N)}{N}} = \sum_\limits{n=0}^{N-1} x_n e^{-i \frac{2\pi n k}{N}} e^{-i2 \pi n} = \sum_\limits{n=0}^{N-1} x_n e^{-i \frac{2\pi n k}{N}} = X_k$ (since $e^{-i2 \pi n} = 1$, for any integer $n$).

**Example:**

Let's take a time varying signal of frequency 1 Hz: $f(t) = sin(2 \pi t)$. It has amplitude $A = 1$. Frequency 1 Hz means the signal has one complete cycle in one second. We want to do DFT of the signal.

Let's collect $N = 8$ discrete points from the signal; this means we have 8 samples or value of the signal at 8 time points. The *sampling rate* or the *sampling frequency* is 8 per sec, here. We have the values of the signal (or the displacement of the wave) at $t = 0, 1/8, 2/8, 3/8......, 7/8$ sec, respectively. Note that, the end point $t = 8/8$ or 1 sec is not considered here so that the next set of samples may start from zero again. As we place such sets of points, the periodicity of the signal is retained.

So, here, \\
$x_0 = f(0) = sin(2 \pi \times 0) = 0$, \\
$x_1 = f(1) = sin(2 \pi \times \frac{1}{8}) = 0.707$, \\
$x_2 = f(2) = sin(2 \pi \times \frac{2}{8}) = 1$, \\
$x_3 = f(3) = sin(2 \pi \times \frac{3}{8}) = 0.707$, \\
$x_4 = f(4) = sin(2 \pi \times \frac{4}{8}) = 0$, \\
$x_5 = f(5) = sin(2 \pi \times \frac{5}{8}) = -0.707$, \\
$x_6 = f(6) = sin(2 \pi \times \frac{6}{8}) = -1$, \\
$x_7 = f(7) = sin(2 \pi \times \frac{7}{8}) = -0.707$, \\

The formula for DFT: $X_k = \sum_\limits{n=0}^{N-1} x_n e^{-i \frac{2\pi n k}{N}}$, ($k$ is in the range $[0, N-1]$).

We now calculate $X_0, X_1, X_2...X_7$.

For $k = 0$, we find $X_0 = 0$. \\
For $k = 1$, we find $X_1 = -4i$.

Similarly, we calculate all other Fourier transformed values in the frequency domain. We have,

$X_0 = 0, X_1 = 0 - 4i, X_2 = 0, X_3 = 0, X_4 = 0, X_5 = 0, X_6 = 0, X_7 = 0 + 4i$.

Note that the DFT values are complex numbers $A+iB$ because of the complex exponential factors. So, we plot the magnitudes $\sqrt{A^2 + B^2}$ at each frequency values.

Now what are the frequency values?

Frequency values start from 0 Hz. Frequency resolution is given by, $\Delta \nu = \frac{R}{N}$, where $R$ is the sampling frequency or sampling rate. It is the value of the sample points collected per second for the time varying signal. Here, $R = 8$. So, the frequencies are,

$\nu_k = \frac{k R}{N}$, for $k = 0, 1, 2,..., N-1$.

```python
# Script for doing DFT.

import numpy as np

# Define the time scale. We take 1 sec time and collect sample at interval of 1/8 sec.

t = np.arange(0, 1, 1/8)
print('The time points at which samples are collected:', t)


# Values of the signal at the time points. We define a 1 Hz sine wave signal.

x = np.sin(2 * np.pi * t)
print('The value of the signal or wave of specific frequency at the collected time points:', x)

# N = total number of samples = length of the time point array.
N = len(t)
print('The total number of samples = ', N)

n = np.arange(N)
print('The array of indices of samples (n) in time domain: ', n)


# Calculate DFT.

# Define an empty list. This will contain all the Fourier components or all the DFT values.
# Basically X = [X_0, X_1, X_2, ..., X_N-1] = [X_k], where k = [0, N-1].

X = []

for k in range(N):   # The loop runs over all the k values: [0, N-1].
    exp_factor = np.exp(-1j * 2 * np.pi * n * k / N)
    # print(exp_factor)

    X_k = np.sum(x * exp_factor)   # this is the value of DFT for a specific k value.
    # print(np.round(X_k))   # Get the complex values of the DFT X_k, with round off.
    # print(np.round(abs(X_k)))   # Get the absolute value of the complex number of DFT X_k, with round off.
    X.append(np.round(abs(X_k)))   # Append the round off of the absolute value of the DFT value X_k, into the list X.

print('List of DFT values: ', X)  # This is a list, contains the DFT values for each k i.e., X_k.
X = np.array(X)   # Converting the list to a numpy array.
print('Numpy array of DFT values: ', X)


# Get the frequencies. frequency is given by, nu_k = (k * R) / N, where k = [0, N-1].
R = 8
freq = np.array([(k * R) / N for k in range(N)])
print('The array of frequency values (in Hz): ', freq)


# For plotting.

import matplotlib.pyplot as plt

plt.plot(freq, X, 'o')   # Plot the X array (The DFT values) vs frequency array.

# We draw a vertical line at each non-zero X_k value, the height of the vertical line is equal to the magnitude of X_k.
for i in range(N):
    if X[i] != 0:
        plt.vlines(freq[i], 0, X[i])
plt.axhline(0, color = 'black', linestyle = '--')  # Draw a horizontal dashed line at y = 0.
plt.show()
```

## Fast Fourier Transform (FFT) and Inverse FFT

In practice, FFT is done using an efficient algorithm of Cooley and Turkey. The essential idea is to recursively divide DFT into smaller DFTs which bring downs the needed computation time drastically. A standard DFT scales as O($N^2$), where as the FFT scales as O($N \log{N}$). FFT is one of the most beautiful and useful algorithms in the entire computational science.

Some important functions in Scipy for doing FFT: \\
**fft()** : Discrete Fourier transform of a list of numbers. \\
**ifft()** : Inverse FFT. \\
**fftfreq()** : To create frequecy domain. \\

The FFT is given by, \\

$X[k] = \sum_\limits{n = 0}^{N-1} x[n] e^{-i\frac{2 \pi nk}{N}}$ . \\

It can be seen from the definition, with $k = 0$, \\

$X[0] = \sum_\limits{n = 0}^{N-1} x[n]$  . \\

If the total number of samples, $N$ is even, the elements $X[0], X[1],...,X[\frac{N}{2} - 1]$ contain the positive frequency terms, and the elements $X[\frac{N}{2}], X[\frac{N}{2} + 1]...., X[N-1]$ contain the negative frequency terms, in order of decreasing negative frequency.

For $N$ odd, the elements $X[1], X[2],...,X[\frac{N-1}{2}]$ contain the positive frequency terms, and the elements $X[\frac{N+1}{2}],..., X[N-1]$ contain the negative frequency terms, in order of decreasing negative frequency.

For real valued $x[n]$, the values of $X[k]$, for positive frequencies are the complex conjugate of the values of $X[k]$ for negative frequencies (because the spectrum is symmetric). Typically only the FFT corresponding to the positive frequencies are plotted.

```python
# Script for doing FFT.

import numpy as np
from scipy.fft import fft, ifft, fftfreq

# Define the time scale.

t = np.arange(0, 1, 1/8)
print('Array of time points: ', t)

# Values of the signal at the time points. We define a 1 Hz sine wave signal.

x = np.sin(2 * np.pi * 1 * t)
print('Value of the signal at the time points: ', x)

# FFT.

X = fft(x)
print('Array of FFT values (round off): ', np.round(X))
print('Array of absolute values of the FFT values (round off): ', np.round(abs(X)))

# IFFT.

IFFT = ifft(X)   # The ifft() returns the signal values (in complex form) back in the time domain.
print('Array of IFFT values (round off upto 2 decimel places)', np.round(IFFT, 2))


# Frequencies.

N = len(t)
dt = 1/8

freq = fftfreq(N, dt)
print('Array of frequencies (in Hz): ', freq)


# For plotting.

import matplotlib.pyplot as plt
plt.plot(freq, abs(X), 'o')

# We draw a vertical line at each non-zero X_k value, the height of the vertical line is equal to the magnitude of X_k.
for i in range(N):
    if X[i] != 0:
        # print(freq[i])
        plt.vlines(freq[i], 0, abs(X[i]))
plt.axhline(0, color = 'black', linestyle = '--')  # Draw a horizontal dashed line at y = 0.
plt.show()
```

**One-sided frequency spectrum**

We can use **rfft()** function to get the frequency spectrum only for the positive frequency (as the spectrum is symmetric).

We will do FFT only for the positive frequency using **rfft()** function next. The signal is given by,

$x(t) = \sin(2 \pi \times 50 t) + \frac{1}{2} \sin(2 \pi \times 100 t)$  .

```python
# Script for one-sided FFT (only for positive frequency spectrum).

# Script for doing FFT.

import numpy as np
from scipy.fft import rfft, rfftfreq

# Define the time scale.

t = np.arange(0, 2, 1/1000)
print('Array of time points: ', t)

# Values of the signal at the time points. We define a 1 Hz sine wave signal.

x = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 100 * t)
print('Value of the signal at the time points: ', x)

# FFT.

X = rfft(x)
X_mag = abs(X)   # Magnitude of FFT values.
print('Array of FFT values (round off): ', np.round(X))
print('Array of absolute values of the FFT values (round off): ', np.round(X_mag))


# Frequencies.

N = len(t)
dt = 1/1000

freq = rfftfreq(N, dt)
print('Array of frequencies (in Hz): ', freq)


# For plotting.

import matplotlib.pyplot as plt
plt.plot(freq, X_mag / np.max(X_mag))   # We plot magnitude of FFT values normalized with the maximum value vs freq.
plt.axhline(0, color = 'black', linestyle = '--')  # Draw a horizontal dashed line at y = 0.
plt.title('FFT spectrum', color = 'r')
plt.xlabel('Frequency')
plt.ylabel('FFT value (normalized)')
plt.grid()
plt.show()
```

We see above the normalized FFT spectrum of the mixed signal composed of two sine waves, of frequencies 50 Hz and 100 Hz. The proportion of the amplitudes of the two signal components is evident from the plot.

# Exercise: Calculate value of $\pi$

**Calculate the value of $\pi$:** It is the ratio of the circumference and diameter of a circle. As circle becomes bigger or smaller, this ratio become constant. Consider a circle having radius $r$. Archimidis first obtain the value of $\pi$ from two regular polygons, one inside the circle, and the other outside the circle. He calculated the ratio $\frac{circumference}{diameter}$ for these two polygons.

Consider both the polygons have $n$ sides, the side of the inner polygon is $a_{in}$ and that of the outer polygon is $a_{out}$.

So, for both the polygons, the exterior angle is $2\pi / n$.

A simple picture tells us, for inner polygon, $\sin(\pi/n) = a_{in}/2r$, and for outer polygon, $\tan(\pi/n) = a_{out}/2r$.

Now the ratio $\frac{circumference}{diameter} = \frac{a_{in}}{2r} = \frac{a_{out}}{2r}$.

This implies, $\frac{circumference}{diameter} = n \sin(\pi/n)$, for inner polygon, and

$\frac{circumference}{diameter} = n \tan(\pi/n)$, for outer polygon.

Here, for circle, inner and outer polygons, the diameter is $2r$.

But, circumference of circle is greater than the circumference of inner polygon, and shorter than the circumference of outer polygon. So,

$n \sin(\pi/n) < \pi_{estimated} < n \tan(\pi/n)$.

So, as we increase the number of sides of polygons, it approaches towards circle. So as we increase value of $n$ (the number of sides of the inner/outer polygons), $\pi_{estimated}$ approaches towards the actual value of $\pi$. So, as $n$ increases, $\pi_{estimated}$ matches with more decimel places of actual $\pi$.

```python
import numpy as np

LL = []
UL = []

for n in range(1, 10000):
    lower_limit = n * np.sin(np.pi/n)
    upper_limit = n * np.tan(np.pi/n)
    LL.append(lower_limit)
    UL.append(upper_limit)

LL = np.array(LL)
UL = np.array(UL)

print(LL)
print(UL)
```
